# Medidas de dispersión:

## Medidas de variabilidad

Las medidas de variabilidad son una medidas estadísticas que indican la variabilidad de un conjunto de datos. Es decir, las medidas de variabilidad miden la dispersión de una serie de datos.

**Definición: Varianza Muestral**

La varianza muestral, denotada por $s^2$ está dada por
$$
s^2=\frac{\sum(x_i-\bar{x})^2}{n-1}
$$

mientras que la desviación estándar (denotada por $s$) corresponde a la raíz de la varianza.


Ejemplo:

De la base tips, calcular la varianza y la desviación estándar de la variable tip

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base=sns.load_dataset('tips')
```

```python
# Desviación estándar
std=base['tip'].std()
```

```python
# Varianza
varianza=base['tip'].var()
```

```python
print(f' Varianza :{round(varianza,2)},   Desviación Estándar: {round(std,2)}')
```

### Definición: (Coeficiente de variación)

El coeficiente de variación se define como la  desviación estándar dividido por la media ($\bar{X}$) y multiplicado por 100. Es decir

$$
CV=\frac{s}{\bar{x}}\cdot100
$$

**Observación:**

 El coeficiente de variación, nos permite determinar la dispersión de datos con respecto a su media.


# Medidas de forma

Las medidas de forma nos permiten identificar la forma y las carecterísticas de la distribución.

## Curtosis:

La curtosis es una medida estadística indica el grado de concentración de una distribución alrededor de su media.

Existen tres tipos principales de curtosis.

1. **Leptocúrtica** (curtosis mayor a 3)

la distribución es muy apuntada, es decir, los datos están muy concentrados alrededor de la media, pero con valores extremos.

2. **Mesocúrtica** (curtosis igual a 3)

la curtosis de la distribución es equivalente a la curtosis de la distribución normal.

3. **Platicúrtica** (curtosis menor a 3)

la concentración en torno a la media es baja. Los datos están más dispersos, con menores valores extremos.

**Ejemplo**

Estudiaremos la curtosis de la variable total_bill de la base de datos tips.

```python
curtosis=base['total_bill'].kurt()
```

```python
print(f' Curtosis: {round(curtosis,2)}')
print('Existe una gran dispersión de datos de datos en el conjunto de datos perteneciente a la variable total bill')
```

Observación:

Para una muestra de tamaño $n$, se define la curtosís como
$$
K=\frac{n(n+1)}{(n-1)(n-2)(n-3)}\sum \left(\frac{(x_i-\bar{x})^4}{s^4}\right)-\frac{3(n-1)^2}{(n-2)(n-3)}
$$
con
- $\bar{x}$ la media
-  $s$ la desviación estándar


## Asimetría:

La asimetría es una medida que indica el grado de simetría (o asimetría) de una distribución respecto a su media. Es decir, la asimetría es un parámetro estadístico que sirve para determinar cuánto de simétrica (o asimétrica) es una distribución sin necesidad de representarla gráficamente.

Tipos de asimetría:

1. **Asimetría positiva** (skewness >0)

En este caso, la distribución de datos se concentran principalmente en el lado izquierdo de la media (la mayoría de los datos, están por debajo de la media)

2. **Asimetría negativa** ( skewness $<0$)

Los datos se concentran en el lado derecho de la media (la mayoría de los valores de la muestra, están por encima de está)

3. **Simetría:** ( skewness $\sim 0$)

 la distribución tiene el mismo número de valores a la izquierda que a la derecha de la media.

#### Definición:

Un experimento aleatorio es aquel que puede producir resultados diferentes, aún cuando se repita de la misma manera.


#### Definición (Espacio Muestral)

Al conjunto de todos los resultados posibles de un experimento aleatorio se le llama *espacio muestral* del experimento. El espacio muestral se denota por $S$

Ejemplo:

El experimento más simple al que se le aplica la probabilidad es uno con dos posibles resultados, tal experimento consiste en lanzar una moneda al aire y ver si al caer el resultado es "cara o cruz".

El espacio muestral de este experimento es $\mathcal{S}=\{\mathrm{cara},\mathrm{cruz}\}$.

Supongamos que realizamos el experimento de lanzar mil veces una moneda al aire para determinar la probabilidad de que salga "cara o cruz":

### Definición (Evento)

Un evento es cualquier recopilación de resultados contenidos en el espacio muestral $\mathcal{S}$. Un evento es simple si consiste en exactamente un resultado y compuesto, si consiste en más de un resultado.

Ejemplo:

Consideramos el experimento de lanzar mil veces la moneda al aire para determinar si sale cara o cruz.

Se define el evento
$S_1$={Obtener cara en los 3 primeros lanzamientos}
Notar que

$$
S_1\subset\mathcal S
$$

### Probabilidad frecuentista (Importante)

Consideremos un espacio muestral $\mathcal{S}$ con una cantidad finita de elementos.

Consideremos el evento $E$ asociado a este espacio muestral, entonces la probabilidad de que ocurra el evente $E$ se define como
$$
P(E)=\frac{\mathrm{total\,resultados\,posibles\, del\,evento\,E}}{\mathrm{resultados\, totales}}
$$

Ejemplo:

Consideramos el experimento lanzar un dado:

La probabilidad de obtener un número par es

$$
P(N)=\frac{3}{6}
$$

### Ejemplo:

Considerar el experimento "lanzar una moneda al aire".

En este caso, el espacio muestral viene dado por $\mathcal{S}$ ={cara,cruz }. Determinemos la probabilidad de que sucedan los siguientes eventos

1. $E_1=$ { al lanzar la moneda, se obtenga cara}.

En este caso se tiene que, (como el número de casos totales es 2)

### Propiedades:

Sea $\mathcal{S}$ y  sea $A$ un evento tal que $A \subset \mathcal{S}$.

Entonces la probabilidad de que suceda el evento $A,\quad P(A)$, cumple las siguientes propiedades:

1. $0\leq P(A)\leq 1$
2. $P(\mathcal{S})=1$
3. $P(\emptyset)=0$
4. Si dos eventos $A$ y $B$ son excluyentes, es decir $A\cap B=\emptyset$, entonces $P(A\cap B)=P(A)+P(B)$
5. $P(A)=1-P(A^C)$, de acá se puede deducir que $P(A)+P(A^C)=1$ (importante)
6. $P(A\cup B)=P(A)+P(B)-P(A\cap B)$

## Probabilidad condicionada

(más  importante que veremos hoy)

Es muy común tener que conocer la probabilidad de que suceda un evento, cuando ya se sabe que el otro evento sucedió. A este evento se le conoce como probabilidad condicionada, que se define como
$$
P(A|B)=\frac{P(A\cap B)}{P(B)}
$$
donde $P(A|B)$ se lee como: "probabilidad de que suceda el evento $A$, dado que el evento $B$ ya sucedio"

## Ejemplo 1:

De un grupo de 70 estudiantes, se conoce que 35 alumnos aprobaron álgebra, 30 cálculo y 5 alumnos ambas materias. Si escogemos un alumno al azar, calcular la probabilidad de que:

1. Aprueba Cálculo, dado que aprobó álgebra
2. Apruebe alguna de las materias

Sol:

Consideremos los eventos $A=$ {individuo aprueba álgebra} y $C=$ {individuo aprueba cálculo}.

 Notar que podemos expresar el grupo de los individuos que no aprueban álgebra como $A^c$ y a los individuos que no aprueban cálculo como $C^c$. Así

1.  En primer lugar debemos calcular $P(C| A)$ y para esto utilizamos el teorema de probabilidad condicional y tenemos:

$$
P(C|A)=\frac{P(C\cap A)}{P(A)}
$$

2. Aprobar alguna de las asignaturas (es decir aprobar $A$ o $C$). En este caso, debemos calcular

$$
P(A\cup C)=P(A)+P(C)-P(A\cap C)
$$
