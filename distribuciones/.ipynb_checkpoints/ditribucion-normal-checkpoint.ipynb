### 📌 1. ¿Qué es una Distribución de Probabilidad

Una **distribución de probabilidad** describe cómo se distribuyen los valores de una variable aleatoria. Puede ser:

- **Discreta*: Valores finitos o contables (e.g., lanzamiento de un dad).
- **Continua*: Valores infinitos dentro de un rango (e.g., altura de persona).

---

### 📈 2. ¿Qué es la Distribución Norma?

La **distribución normal**, también conocida como **distribución gaussiana**, es una distribución de probabilidad continua caracterizada por su forma de campana simétrica. Depende de dos parámetros:

- **Media (μ)*: Ubicación del centro de la distribucón.
- **Desviación estándar (σ)*: Medida de la dispersión de los datos alrededor de la media.

#### 📐 Función de Densidad de Probabilidad (PF)

La función de densidad de una variable aleatoria continua \( X \) con distribución norma es:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

Donde:

- $ \mu $: Media.
- $ \sigma $: Desviación estádar.
- $ x $: Valor de la variable aleatoria.

---

### 📊 3. Variables que siguen una Distribución Normal

Muchas variables en la naturaleza y en contextos sociales tienden a seguir una distribución normal, como:

- Altura y peso de indivduo.
- Presión arterial.
- Errores de medición.
- Resultados de exámenes estandarizado.
- Tiempo de viaje al trabajo.

---

### 🧪 4. Verificar la Normalidad en Python

Para evaluar si un conjunto de datos sigue una distribución normal, se pueden utlizar:

#### 📉 Visualizaciones

- **Histogama**: Observa la forma general de la distrbución.
- **Gráfico Q-Q (Quantile-Quantle)**: Compara los cuantiles de los datos con los de una distribución normal.

#### 🧪 Pruebas Estadísticas

- **Prueba de ShapiroWilk**: Evalúa la hipótesis nula de que los datos provienen de una distribució normal.
- **Prueba de Kolmogorov-Smrnov**: Compara la distribución empírica de los datos con una distribución de referencia.

---

### 🔄 5. Transformar Variables hacia la Nomalidad

Si los datos no siguen una distribución normal, se pueden aplicar transformaciones para aprximarlos:

- **Transformación logaritmica**: Útil para datos sesgados positivamente.
- **Transformación de Box-Cox**: Aplica una transformación de potencia para estabilizar la varianza y hacer que los datos se asemejen a una distribucin normal.
- **Transformación de Yeo-Jhnson**: Similar a Box-Cox, pero puede aplicarse a datos con valores cero o negativos.

---

### ⚠️ 6. Problemas Asociados con la Nomalidad

No todos los conjuntos de datos siguen una distribución normal. Las desviaciones pueden deberse a:

- **Sesgo (Skeness)**: Asimetría en la disribución.
- **Curtosis (Kurosis)**: Presencia de colas más pesadas o ligeras que las de una distribucin normal.
- **Ouliers**: Valores atípicos que afectan la forma de la distribución.

Es crucial identificar y tratar estos problemas antes de aplicar modelos que asumen nrmalidad.

---

La distribución normal es fundamental en estadística y ciencia de datos debido a su prevalencia en fenómenos naturales y su papel central en muchos métodos estadísticos. Comprender sus propiedades y cómo verificar la normalidad en los datos es esencial para aplicar correctamente técnicas de análisis  modelado.

## 2. 🐍 Código para Distribución Normal

### 🧰 Librerías necesarias

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, normaltest
````

### 🔢 Paso 1: Generar datos con distribución normal

````python
np.random.seed(42)  # para reproducibilidad
datos = np.random.normal(loc=50, scale=10, size=1000)
````

📌 Explicación:

`loc=50`: media (μ).

`scale=10`: desviación estándar (σ).

`size=1000`: cantidad de datos.

### 📊 Paso 2: Visualizar la distribución


````python
sns.histplot(datos, bins=30, kde=True, color="skyblue")
plt.title("Distribución Normal Simulada")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
````

También podemos agregar la curva teórica de la distribución normal:

````python
x = np.linspace(min(datos), max(datos), 100)
plt.plot(x, norm.pdf(x, loc=50, scale=10), label="Distribución Normal Teórica", color='red')
sns.histplot(datos, kde=True, stat='density', bins=30, color='skyblue')
plt.legend()
plt.title("Datos + Curva Normal Teórica")
plt.xlabel("Valores")
plt.show()
````

### 🧪 Paso 3: Test de normalidad (Shapiro-Wilk)

````python
stat, p = shapiro(datos)
print("Estadístico:", round(stat,2))
print("Valor p:", round(p,2))

if p > 0.05:
    print("✅ Los datos parecen seguir una distribución normal (no se rechaza H0).")
else:
    print("❌ Los datos NO parecen seguir una distribución normal (se rechaza H0).")
````

📌 **Interpretación de los Tests de Normalidad**

- **Hipótesis nula (H₀):** Los datos siguen una distribución normal.

- Si $ p > 0.05 $, no se rechaza H₀ → ✅ **Los datos se consideran normales**.

- Si $ p \leq 0.05 $, se rechaza H₀ → ❌ **Los datos no siguen una distribución normal**.

📌 **Fórmula de la distribución normal:**

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

Donde:
- $\mu $ es la media
- $ \sigma $ es la desviación estándar
- $ x $ es el valor de la variable aleatoria


## 3. 🐍 Código para Evaluación de Normalidad

### 🔢 Generar datos con distribución normal

````python
np.random.seed(42)
datos = np.random.normal(loc=50, scale=10, size=1000)
````

### 📊 Visualización previa

````python
sns.histplot(datos, kde=True, bins=30, color="skyblue")
plt.title("Distribución de Datos Simulados")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
````

### 🧪 1. Test de Shapiro-Wilk

Ideal para muestras pequeñas/medianas (n < 5000)


````python
stat_shapiro, p_shapiro = shapiro(datos)
print("Shapiro-Wilk Test:")
print(f"Estadístico: {stat_shapiro:.4f} - Valor p: {p_shapiro:.4f}")

if p_shapiro > 0.05:
    print("✅ Shapiro-Wilk: Normalidad NO rechazada.")
else:
    print("❌ Shapiro-Wilk: Se rechaza normalidad.")
````

### 📈 2. Test de Kolmogorov-Smirnov

Compara la muestra con una **distribución teórica** (en este caso, normal con media y desviación estándar de la muestra).

````python
stat_kstest, p_kstest = kstest(datos, 'norm', args=(np.mean(datos), np.std(datos)))
print("\nKolmogorov-Smirnov Test:")
print(f"Estadístico: {stat_kstest:.4f} - Valor p: {p_kstest:.4f}")

if p_kstest > 0.05:
    print("✅ Kolmogorov-Smirnov: Normalidad NO rechazada.")
else:
    print("❌ Kolmogorov-Smirnov: Se rechaza normalidad.")
````

### 📉 3. Test de D’Agostino y Pearson

Combina **asimetría y curtosis**. Bueno para muestras más grandes.

````python
stat_normaltest, p_normaltest = normaltest(datos)
print("\nD’Agostino y Pearson (normaltest):")
print(f"Estadístico: {stat_normaltest:.4f} - Valor p: {p_normaltest:.4f}")

if p_normaltest > 0.05:
    print("✅ Normaltest: Normalidad NO rechazada.")
else:
    print("❌ Normaltest: Se rechaza normalidad.")
````

### ✅ Recomendaciones

| Test                         | Recomendado para             | Fortalezas                              |
|-----------------------------|------------------------------|-----------------------------------------|
| **Shapiro-Wilk**            | n < 5000                     | Alta potencia con pocas observaciones   |
| **Kolmogorov-Smirnov**      | Comparación con dist. teórica| Fácil de usar pero menos potente        |
| **Normaltest (D’Agostino)** | n > 20                       | Evalúa simetría y curtosis              |


### 🧪 Comparación de Pruebas de Normalidad: ¿Qué pasa cuando no coinciden?

````python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, normaltest

# 🔢 Generamos una muestra sesgada positivamente (log-normal)
np.random.seed(42)
data = np.random.lognormal(mean=3.8, sigma=0.4, size=100)

# 📊 Visualización
sns.histplot(data, kde=True, bins=30, color="salmon")
plt.title("Distribución Sesgada (Log-Normal)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# 🧪 Test de Shapiro-Wilk
stat_shapiro, p_shapiro = shapiro(data)
print(f"Shapiro-Wilk: Estadístico = {stat_shapiro:.4f}, p = {p_shapiro:.4f}")

# 📈 Test de Kolmogorov-Smirnov (ajustado a N(μ, σ))
mu, sigma = np.mean(data), np.std(data)
stat_ks, p_ks = kstest(data, 'norm', args=(mu, sigma))
print(f"Kolmogorov-Smirnov: Estadístico = {stat_ks:.4f}, p = {p_ks:.4f}")

# 📉 D’Agostino y Pearson (normaltest)
stat_nt, p_nt = normaltest(data)
print(f"D’Agostino-Pearson: Estadístico = {stat_nt:.4f}, p = {p_nt:.4f}")
````

### 📋 Posible salida esperada

```
Shapiro-Wilk: Estadístico = 0.9842, p = 0.1675   ✅
Kolmogorov-Smirnov: Estadístico = 0.0862, p = 0.0501  ⚠️
D’Agostino-Pearson: Estadístico = 7.4321, p = 0.0241 ❌
```

### 📌 Interpretación

| Test                        | Resultado      | ¿Rechaza Normalidad? | Comentario                                                                 |
|----------------------------|----------------|-----------------------|---------------------------------------------------------------------------|
| **Shapiro-Wilk**           | p > 0.05       | ❌ No                 | Muy tolerante a pequeñas desviaciones.                                    |
| **Kolmogorov-Smirnov**     | p ≈ 0.05       | ⚠️ Dudoso             | Cercano al umbral. Menos sensible a sesgo, pero útil para distribución teórica. |
| **D’Agostino-Pearson**     | p < 0.05       | ✅ Sí                 | Detectó sesgo y curtosis, lo que otras pruebas ignoraron.                |

### ⚠️ ¿Por qué las pruebas no coinciden?

#### 🎯 Causas

- **Shapiro-Wilk**: Es muy potente para detectar *ligeras* desviaciones en muestras pequeñas/medianas, pero **menos sensible al sesgo y curtosis leve**.
- **Kolmogorov-Smirnov**: Compara con una **distribución teórica** (en este caso N(μ,σ)) y puede pasar por alto detalles como asimetría si los parámetros son "ajustados".
- **D’Agostino-Pearson**: Analiza explícitamente **simetría (skewness)** y **forma de las colas (kurtosis)**. Es más riguroso y suele detectar problemas de normalidad donde otras no.


#### ✅ Conclusión

Cuando las pruebas **no concuerdan**:

- **Visualiza primero** (histograma + Q-Q plot).
- **Confía más en D’Agostino-Pearson o Shapiro-Wilk** según el tamaño de muestra y naturaleza del sesgo.
- Considera el contexto: ¿es necesario asumir normalidad o puedes usar métodos no paramétricos?
