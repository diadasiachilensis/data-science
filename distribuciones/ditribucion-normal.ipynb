### 📌 1. ¿Qué es una Distribución de Probabilidad

Una **distribución de probabilidad** describe cómo se distribuyen los valores de una variable aleatoria. Puede ser:

- **Discreta*: Valores finitos o contables (e.g., lanzamiento de un dad).
- **Continua*: Valores infinitos dentro de un rango (e.g., altura de persona).

---

### 📈 2. ¿Qué es la Distribución Norma?

La **distribución normal**, también conocida como **distribución gaussiana**, es una distribución de probabilidad continua caracterizada por su forma de campana simétrica. Depende de dos parámetros:

- **Media (μ)*: Ubicación del centro de la distribucón.
- **Desviación estándar (σ)*: Medida de la dispersión de los datos alrededor de la media.

#### 📐 Función de Densidad de Probabilidad (PF)

La función de densidad de una variable aleatoria continua \( X \) con distribución norma es:

\[
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}\]

Donde:

- \( \mu\): Media.
- \( \sigma\): Desviación estádar.
- \( x\): Valor de la variable aleatoria.

---

### 📊 3. Variables que siguen una Distribución Normal

Muchas variables en la naturaleza y en contextos sociales tienden a seguir una distribución normal, como:

- Altura y peso de indivduo.
- Presión arterial.
- Errores de medición.
- Resultados de exámenes estandarizado.
- Tiempo de viaje al trabajo.

---

### 🧪 4. Verificar la Normalidad en Python

Para evaluar si un conjunto de datos sigue una distribución normal, se pueden utlizar:

#### 📉 Visualizaciones

- **Histogama**: Observa la forma general de la distrbución.
- **Gráfico Q-Q (Quantile-Quantle)**: Compara los cuantiles de los datos con los de una distribución normal.

#### 🧪 Pruebas Estadísticas

- **Prueba de ShapiroWilk**: Evalúa la hipótesis nula de que los datos provienen de una distribució normal.
- **Prueba de Kolmogorov-Smrnov**: Compara la distribución empírica de los datos con una distribución de referencia.

---

### 🔄 5. Transformar Variables hacia la Nomalidad

Si los datos no siguen una distribución normal, se pueden aplicar transformaciones para aprximarlos:

- **Transformación logaritmica**: Útil para datos sesgados positivamente.
- **Transformación de Box-Cox**: Aplica una transformación de potencia para estabilizar la varianza y hacer que los datos se asemejen a una distribucin normal.
- **Transformación de Yeo-Jhnson**: Similar a Box-Cox, pero puede aplicarse a datos con valores cero o negativos.

---

### ⚠️ 6. Problemas Asociados con la Nomalidad

No todos los conjuntos de datos siguen una distribución normal. Las desviaciones pueden deberse a:

- **Sesgo (Skeness)**: Asimetría en la disribución.
- **Curtosis (Kurosis)**: Presencia de colas más pesadas o ligeras que las de una distribucin normal.
- **Ouliers**: Valores atípicos que afectan la forma de la distribución.

Es crucial identificar y tratar estos problemas antes de aplicar modelos que asumen nrmalidad.

## 2. 🐍 Código

### 🧰 Librerías necesarias

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, normaltest
```

### 🔢 Paso 1: Generar datos con distribución normal

```python
np.random.seed(42)  # para reproducibilidad
datos = np.random.normal(loc=50, scale=10, size=1000)
```

📌 Explicación:

`loc=50`: media (μ).

`scale=10`: desviación estándar (σ).

`size=1000`: cantidad de datos.

### 📊 Paso 2: Visualizar la distribución


```python
sns.histplot(datos, bins=30, kde=True, color="skyblue")
plt.title("Distribución Normal Simulada")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

También podemos agregar la curva teórica de la distribución normal:

```python
x = np.linspace(min(datos), max(datos), 100)
plt.plot(x, norm.pdf(x, loc=50, scale=10), label="Distribución Normal Teórica", color='red')
sns.histplot(datos, kde=True, stat='density', bins=30, color='skyblue')
plt.legend()
plt.title("Datos + Curva Normal Teórica")
plt.xlabel("Valores")
plt.show()
```
