## 2.  C贸digo para Distribuci贸n Normal

### О Librer铆as necesarias

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, normaltest
````

###  Paso 1: Generar datos con distribuci贸n normal

````python
np.random.seed(42)  # para reproducibilidad
datos = np.random.normal(loc=50, scale=10, size=1000)
````

 Explicaci贸n:

`loc=50`: media (渭).

`scale=10`: desviaci贸n est谩ndar ().

`size=1000`: cantidad de datos.

###  Paso 2: Visualizar la distribuci贸n


````python
sns.histplot(datos, bins=30, kde=True, color="skyblue")
plt.title("Distribuci贸n Normal Simulada")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
````

Tambi茅n podemos agregar la curva te贸rica de la distribuci贸n normal:

````python
x = np.linspace(min(datos), max(datos), 100)
plt.plot(x, norm.pdf(x, loc=50, scale=10), label="Distribuci贸n Normal Te贸rica", color='red')
sns.histplot(datos, kde=True, stat='density', bins=30, color='skyblue')
plt.legend()
plt.title("Datos + Curva Normal Te贸rica")
plt.xlabel("Valores")
plt.show()
````

### И Paso 3: Test de normalidad (Shapiro-Wilk)

````python
stat, p = shapiro(datos)
print("Estad铆stico:", round(stat,2))
print("Valor p:", round(p,2))

if p > 0.05:
    print(" Los datos parecen seguir una distribuci贸n normal (no se rechaza H0).")
else:
    print(" Los datos NO parecen seguir una distribuci贸n normal (se rechaza H0).")
````

 **Interpretaci贸n de los Tests de Normalidad**

- **Hip贸tesis nula (H):** Los datos siguen una distribuci贸n normal.

- Si $ p > 0.05 $, no se rechaza H   **Los datos se consideran normales**.

- Si $ p \leq 0.05 $, se rechaza H   **Los datos no siguen una distribuci贸n normal**.

 **F贸rmula de la distribuci贸n normal:**

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

Donde:
- $\mu $ es la media
- $ \sigma $ es la desviaci贸n est谩ndar
- $ x $ es el valor de la variable aleatoria


## 3.  C贸digo para Evaluaci贸n de Normalidad

###  Generar datos con distribuci贸n normal

````python
np.random.seed(42)
datos = np.random.normal(loc=50, scale=10, size=1000)
````

###  Visualizaci贸n previa

````python
sns.histplot(datos, kde=True, bins=30, color="skyblue")
plt.title("Distribuci贸n de Datos Simulados")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
````

### И 1. Test de Shapiro-Wilk

Ideal para muestras peque帽as/medianas (n < 5000)


````python
stat_shapiro, p_shapiro = shapiro(datos)
print("Shapiro-Wilk Test:")
print(f"Estad铆stico: {stat_shapiro:.4f} - Valor p: {p_shapiro:.4f}")

if p_shapiro > 0.05:
    print(" Shapiro-Wilk: Normalidad NO rechazada.")
else:
    print(" Shapiro-Wilk: Se rechaza normalidad.")
````

###  2. Test de Kolmogorov-Smirnov

Compara la muestra con una **distribuci贸n te贸rica** (en este caso, normal con media y desviaci贸n est谩ndar de la muestra).

````python
stat_kstest, p_kstest = kstest(datos, 'norm', args=(np.mean(datos), np.std(datos)))
print("\nKolmogorov-Smirnov Test:")
print(f"Estad铆stico: {stat_kstest:.4f} - Valor p: {p_kstest:.4f}")

if p_kstest > 0.05:
    print(" Kolmogorov-Smirnov: Normalidad NO rechazada.")
else:
    print(" Kolmogorov-Smirnov: Se rechaza normalidad.")
````

###  3. Test de DAgostino y Pearson

Combina **asimetr铆a y curtosis**. Bueno para muestras m谩s grandes.

````python
stat_normaltest, p_normaltest = normaltest(datos)
print("\nDAgostino y Pearson (normaltest):")
print(f"Estad铆stico: {stat_normaltest:.4f} - Valor p: {p_normaltest:.4f}")

if p_normaltest > 0.05:
    print(" Normaltest: Normalidad NO rechazada.")
else:
    print(" Normaltest: Se rechaza normalidad.")
````

###  Recomendaciones

| Test                         | Recomendado para             | Fortalezas                              |
|-----------------------------|------------------------------|-----------------------------------------|
| **Shapiro-Wilk**            | n < 5000                     | Alta potencia con pocas observaciones   |
| **Kolmogorov-Smirnov**      | Comparaci贸n con dist. te贸rica| F谩cil de usar pero menos potente        |
| **Normaltest (DAgostino)** | n > 20                       | Eval煤a simetr铆a y curtosis              |


### И Comparaci贸n de Pruebas de Normalidad: 驴Qu茅 pasa cuando no coinciden?

````python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, norm, normaltest

#  Generamos una muestra sesgada positivamente (log-normal)
np.random.seed(42)
data = np.random.lognormal(mean=3.8, sigma=0.4, size=100)

#  Visualizaci贸n
sns.histplot(data, kde=True, bins=30, color="salmon")
plt.title("Distribuci贸n Sesgada (Log-Normal)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# И Test de Shapiro-Wilk
stat_shapiro, p_shapiro = shapiro(data)
print(f"Shapiro-Wilk: Estad铆stico = {stat_shapiro:.4f}, p = {p_shapiro:.4f}")

#  Test de Kolmogorov-Smirnov (ajustado a N(渭, ))
mu, sigma = np.mean(data), np.std(data)
stat_ks, p_ks = kstest(data, 'norm', args=(mu, sigma))
print(f"Kolmogorov-Smirnov: Estad铆stico = {stat_ks:.4f}, p = {p_ks:.4f}")

#  DAgostino y Pearson (normaltest)
stat_nt, p_nt = normaltest(data)
print(f"DAgostino-Pearson: Estad铆stico = {stat_nt:.4f}, p = {p_nt:.4f}")
````

- **DAgostino-Pearson**: Analiza expl铆citamente **simetr铆a (skewness)** y **forma de las colas (kurtosis)**. Es m谩s riguroso y suele detectar problemas de normalidad donde otras no.
