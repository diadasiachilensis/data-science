```python
import numpy as np
```

```python
arr=np.linspace(1,10,10, dtype='int8')
arr
```

```python
arr>5
```

```python
#entrega valor booleano a todos los valores dentro del array que cumplen o no cumplen con la condicion
indices_cond=arr>5
indices_cond
```

```python
#muestra una lista con valores que cumplen con la condicion 
arr[indices_cond]
```

```python
#esto tambien entrega el mismo resultado
arr[arr>5]
```

```python
#otra condicional con valores AND
arr[(arr>5)&(arr<9)]
```

```python
arr[arr>5]=99
```

```python
arr[arr==99]
```
