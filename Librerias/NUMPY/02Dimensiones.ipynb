# Dimensiones

indica el numero de dimensiones 
```
ndim 
```



````python
scalar=np.array(42)
print(scalar)
scalar.ndim
````

````python
vector=np.array([1,2,3])
print(vector)
vector.ndim
````

````python
matriz=np.array([[1,2,3],[4,5,6]])
print(matriz)
matriz.ndim
````

#Agregar o eliminar dimensiones 

Agrega el numero de dimensiones que uno desea de una lista 
```
ndim=10
```

axis cuando es cero, se grega una fila y el axis uno se agrega una columna 
```
axis=0 ; axis=1
```






````python
vector=np.array([1,2,3], ndmin=10)
print(vector)
vector.ndim
````

---

con esta función permite expandir el tensor
```
np.expand_dims 
```



````python
expand=np.expand_dims(np.array([1,2,3]), axis=0)
print(expand)
expand.ndim
````

````python
print(vector, vector.ndim)
vector_2=np.squeeze(vector)
````

---
Con esto comprime el numero de dimensiones correcto que no se esten usando 


```
np.squeeze(vector)
```
En el caso de abajo comprime 10 dimensiones a una porque no se estan usando 9 dimensiones 



````python
print(vector, vector.ndim)
vector_2=np.squeeze(vector)
print(vector_2, vector_2.ndim)
````
