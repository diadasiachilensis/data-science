##Merge - Join

*   **Left join:** da prioridad al dataFrame de la izquierda. Trae siempre los datos de la izquierda y las filas en comun con el dataFrame de la derecha.
*   **Right join:** da prioridad al dataFrame de la derecha. Trae siempre los datos de la derecha y las filas en comun con el dataFrame de la izquierda.
*   **Inner join:** Trae solamente aquellos datos que son com√∫n en ambos dataFrame
*   **Outer join:** Trae los datos tanto del dataFrame de la izquierda como el de la derecha incluyendo los datos que comparten ambos.

*   **Concat - Axis 0:** permite combinar dos dataframes a nivel de filas. Crecimiento vertical. 
*   **Concat - Axis 1:** permite combinar dos dataframes a nivel de columnas. La organizacion por columnas no va a ser la misma para ambos dataFrames, por tanto, se crearan valores NaN para rellenar los espacios vacios. Crecimiento horizontal








```python
df1=pd.DataFrame(
{ 'A':['A0','A1','A2','A3',],
  'B':['B0','B1','B2','B3',],
  'C':['C0','C1','C2','C3',],
  'D':['D0','D1','D2','D3',],
 })

df2=pd.DataFrame(
{ 'A':['A4','A5','A6','A7',],
  'B':['B4','B5','B6','B7',],
  'C':['C4','C5','C6','C7',],
  'D':['D4','D5','D6','D7',],
 })
```

#Concat

```python
#se concatenan todos los valores 
#los indices heredan los indices anterior
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2], ignore_index=True)
#respsta los indices y los reordena
```

```python
pd.concat([df1,df2], axis=1)
#amplia de menra horizontal (columnar)
```

#Merge

```python
 izq = pd.DataFrame({
'key' : ['k0', 'k1', 'k2','k3'],
'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']
})

der = pd.DataFrame({
'key' : ['k0', 'k1', 'k2','k3'],
'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']
})
```

```python
izq.merge(der, on='key')
#asi se puede hacer join de varios lados
```

```python
izq.merge(der)
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, on='key')
#falla porque 'key' esta en un solo dattaframe
```

```python
#a partir de lo anterior lo que sirve hacder es
izq.merge(der, left_on='key',right_on='key_2')
```

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
der
#que opasa si una de las filas no tiene un valor no tiene un tipode dato
```

```python
izq.merge(der, left_on='key',right_on='key_2')
#no hace match por lo tanto no desarrolla el index 3 
#solo hace las coincidencias, la interseccion
```

```python
#para un left join
izq.merge(der, left_on='key',right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key',right_on='key_2', how='inner')
#"how='inner'":para es un valor por default
```

```python
#para un right join
izq.merge(der, left_on='key',right_on='key_2', how='right')
```

#Join

```python
izq = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

der =pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3']) 
```

```python
izq.join(der)
#lo hace directamnet de los indices ko, k1 y k2
```

```python
izq.join(der, how='inner')
#solo aparece el k0 y k2 porque es el unico que hace match entre los dos diccionarios
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```

```python
#"how='outer'" respeta los join pero sin importar si los datos estan inscritos
izq.join(der, how='outer')
```
