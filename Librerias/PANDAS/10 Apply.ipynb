```python
import pandas as pd
```

```python
df_books=pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df_books.head(2)
```

```python
def two_times(value):
  return value*2
```

```python
#toma cada una de los valores de la fila que estan en user ratings y le aplica la funcion
#apply es mucho mas rapido que plaicar un bucle for
df_books['User Rating'].apply(two_times)
```

```python
#el Dataframe se le va crear una nueva columna llamda 'Rating_2'
df_books['Rating_2']=df_books['User Rating'].apply(lambda x:x*3)
#ahora vemos la creacion de la columna
df_books
```

```python
df_books['Rating_2']=df_books['User Rating'].apply(lambda x:x*3)
df_books.head()
```

```python
#si quiero ocupar distintas columnas para hacer las operaciones, el apply se loo hago al DaataFrame en general 
#se coloca axis=1 porque les queremos decir que los nombres se hace referencia a la columna 1 'axis=1'
df_books.apply(lambda x:x['User Rating']*2 if x['Genre']=='Fiction' else x['User Rating'], axis=1)
df_books
```
