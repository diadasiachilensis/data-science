```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv',sep=',',header=0)
df_books.head(2)
```

```python
mayor_a_2016=df_books['Year']>2016
```

```python
#se comporta de igual manera que la codncion anterior invocada por la variable
df_books[mayor_a_2016]
```

```python
#esto es lo mismo que lo anterior
df_books[df_books['Year']>2016]
```

```python
genre_fiction=df_books['Genre']=='Fiction'
```

```python
#se pueden juntar dos condiciones
df_books=[genre_fiction & mayor_a_2016]
df_books
```

##Filtrar con negacion 

```python
#filtrar aquellos libros antes del 2017
df_books[~mayor_a_2016]
```
