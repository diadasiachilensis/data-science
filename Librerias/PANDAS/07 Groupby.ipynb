##Grupby

existan ciertos valores que son identicos en algunos valores presentes en un columna. 

se pueden formar en grupos. invocando la funcion groupby.

Captura.PNG

```python
df_books=pd.read_csv('/content/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=',',header=0)
df_books.head(2)
```

```python
#arroja los gurpos de datos presente para cada valor de Author
#se cuenta los registros
df_books.groupby('Author').count()
#fijate autor esta abajo de las columnas, porque es un indeica del DataFrame
```

```python
df_books.groupby('Author').mean()
#fijate autor esta abajo de las columnas, porque es un indeica del DataFrame
```

```python
df_books.groupby('Author').max()
#fijate autor esta abajo de las columnas, porque es un indeica del DataFrame
```

```python
df_books.groupby('Author').min()
#fijate autor esta abajo de las columnas, porque es un indeica del DataFrame
```

```python
#sumas los ratings y los anhos
#fijate autor esta abajo de las columnas, porque es un indeica del DataFrame
df_books.groupby('Author').sum()
```

```python
#si es un indice puedo entrar por loc()
#asi puedo conocer el valor de un indice que en este caso seria William Davis
df_books.groupby('Author').sum().loc['William Davis']
```

```python
#si yo quiero los valores planos, no quiero el Author sea el index
#el index (numeros que estan en la primera columna de la izquierda vuelven)
df_books.groupby('Author').sum().reset_index()
```

```python
#author psa a ser el index
#por author puedo conocer el valor de min y max en cada columna que estoy trabajando 
df_books.groupby('Author').agg(['min','max'])
```

```python
#asi puedo formar columnas nuevas con nuevas funciones de agregacion
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

```python
#se puede ver una llave compuesta con dos index, asi por medio de dos index s epuede conocer los valores de la columna
df_books.groupby(['Author','Year']).count()
```
