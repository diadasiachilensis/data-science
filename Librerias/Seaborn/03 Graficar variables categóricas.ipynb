```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#Importamos las librerias necesarias y llamamos al dataset:

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

#Graficos

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

### Datos de dispersion

```python
plt.figure(figsize=(8,6))
sns.stripplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

###Swarmplot

Muestra la concetracion de los datos

```python
plt.figure(figsize=(8,6))
sns.swarmplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

###Boxplot

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

###Swarmplot y boxplot

muestra la distribucion

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True, color='0', marker='<', alpha=0.5)
plt.show()
```

###Violinplot

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips,x='day', y='total_bill', hue='sex', dodge=True, marker='<', alpha=0.5, palette='pastel',split=True)#split=True, los combina los sexos en un solo violin
plt.show()
```

```python
sns.violinplot(data=tips,x='day', y='total_bill')
```

#Catplot
###kind es una variable que se puede poner cualquier tipo de grafico  

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex', kind='swarm', dodge=True)
plt.show()
```

###col 
Es una variable que puede entregar otro tipo de grafico para constrastar sobre la infroamciion que presenta otra columna

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex', kind='box', dodge=True,col='time')
plt.show()
```
