```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#Importamos las librerias necesarias y llamamos al dataset:

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

#Scatterplot

```python
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day')
plt.show()
```

Se puede agregar otras variables 

```python
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day',style='time')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day',style='time',size='size')
plt.show()
```

se ordena el grafico con `plt.legend`

```python
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day',style='time')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers={
    "Lunch":"D",
    "Dinner":"s"
}
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='day',style='time',markers=markers,size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

#Lineplot

Es una correlacion pero con lineas.
Refleja la relacion entre las cuentas y las propinas

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',style='time',size='size',col='time')
plt.show()
```
