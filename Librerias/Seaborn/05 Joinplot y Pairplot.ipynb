```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#Importamos las librerias necesarias y llamamos al dataset:

```python
tips=sns.load_dataset('tips')
tips.head(2)
```

#Joinplot

Muestra correlacion y distribucion de los datos

```python
sns.jointplot(data=tips,x='total_bill', y='tip')
plt.show()
```

Customizar mas

```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue = 'sex')
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill', y='tip',hue='sex',kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill', y='tip',hue='sex',kind='hist')
plt.show()
```

Agregar más argumentos para poder hacer un mejor análisis


```python
sns.jointplot(data=tips, x='total_bill', y= 'tip', hue='sex', kind= 'hist', marginal_ticks=True, #muestra una pequeña tabla para la gráfica exterior
marginal_kws=dict(bins= 25, fill = True, multiple= 'dodge')); #marginal_kws se le pueden agregar diferentes parámetros,  se pueden modificar los valores de las gráficas secundarias de los ejes
```

#Pairplot

Pairplot correlaciona toda la variable numéricas que hay en el dataset


```python
sns.pairplot(data= tips)
plt.show()
```

Solo muestra una esquina del pairplot, eliminando así, las gráficas repetidas


```python
sns.pairplot(data= tips, corner=True)
```

```python
sns.pairplot(data= tips,hue='sex',palette='coolwarm',corner=True)
plt.show()
```

```python
sns.pairplot(data= tips,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
