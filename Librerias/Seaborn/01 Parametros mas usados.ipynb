```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

#Importamos las librerias necesarias y llamamos al dataset

> Bloque con sangría



```python
tip=sns.load_dataset('tips')
```

```python
tip
```

#Grafico de frecuencia para el total de la cuenta

```python
sns.displot(data=tip, x='total_bill') #saca los datos de la tabla dentro de la variable 'tip', la columna que se extrae es 'total_bill'
```

#Al agregar otra variable, seaborn reconoce que un grafico de frecuencia no es el adecuado por lo que busca el que mejor se adecue.

```python
sns.displot(data=tip, x='total_bill', y='tip')
plt.show()
```

#Agregar una variable de segmentacion mediante hue

```python
sns.displot(data=tip, x='total_bill', y='tip', hue='sex')#hue hace un desglose o un agrupamiento por cualquier otra variable
plt.show()
#con hue resulta el cuadro de hombre y mujer
```

```python
sns.displot(data=tip, x='total_bill', hue='sex')
plt.show()
```

#distribucion por densidad

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind='kde')
plt.show()
#'ked' es la misma distribucion pero por densidad 
```

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind='kde', legend=False) # legend=False, saca la eleyenda
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind='kde', legend=False, palette='dark',alpha=.25)
#alpha hace mas transparente las graficas
plt.show()
```
