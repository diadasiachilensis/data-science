#Set
Es un metodo de seaborn que permite configurar el estilo, fuente y color de las graficas.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Grafica de barplot

```python
sns.barplot(x=['a', 'b', 'c'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=['a', 'b', 'c'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='dark')
sns.barplot(x=['a', 'b', 'c'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid', palette='deep') #pallete y style le da estilo al grafico 
sns.barplot(x=['a', 'b', 'c'], y=[1,2,3])
plt.show()
```

Configura el estilo, la paleta de colores, la fuente y el tamaño de la fuente de la grafica

```python
sns.set(style='darkgrid', palette='dark', font='Verdana', font_scale=2) #font_scale es el tamaño de la fuente | font es el tipo de fuente
sns.barplot(x=['a', 'b', 'c'], y=[1,2,3])
plt.show()
```
