```python
import matplotlib.pyplot as plt
import numpy as np
```

Generar datos

```python
data=np.random.randint(1,50,100)
```

```python
data
```

#Grafico de distribucion y frecuencia

```python
plt.hist(data, bins=10,histtype='bar')#bins ens el numero de barras que va a tener los datos | histtype es el tipo de histograma
plt.show()
```

```python
plt.hist(data, bins=10,histtype='step')#histtype='step -> linea de frecuencia 
plt.show()
```

#Boxplot

```python
plt.boxplot(data, vert=False, patch_artist=True,notch=True) #vert=False se ve de lado horizontal | patch_artist=True, caja rellena | notch=True distribucion de los datos 
plt.show()
```

```python
data=np.append(data,200) #aqui se crea un dato lejano y el grafico lo trata como un outlier 
plt.boxplot(data, vert=False, patch_artist=True,notch=True, showfliers=True)  
plt.show()
```

```python
data=np.append(data,200) #aqui se crea un dato lejano y el grafico lo trata como un outlier 
plt.boxplot(data, vert=False, patch_artist=True,notch=True, showfliers=False) #showfliers=False no te muestra con outliers, solo te muestra la concetracion de los datos 
plt.show()
```

#Scatter

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
area=(30*np.random.rand(N))**2
colors=np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors)#s=area, el tamaño de la burbuja va a indicar una tercera variable | c=colors clasificaicon de la dispersion que se analiza
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors, marker='p', alpha=0.5) #marker, cmabia la forma | alpha=0.5 indica la transparencia 
plt.show()
```

Gracias a la librería GridSpec se puede combinar un gráfico scatter plot de 2 variables con los histogramas de cada una de las 2 variables.
Con esta librería definimos una figura de 4x4 gráficos. Luego compartimos unimos zonas de la cuadrícula de gráficos definidos de manera que:
Para el scatter plot quede una zona de 3x3, para el histograma de x una zona de 1x3 y para el histograma de y una zona de 3x1.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure()
gs = GridSpec(4, 4)

ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

plt.show()
```

```python
fig = plt.figure()
gs = GridSpec(4, 4)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

N = 50 
x = np.random.rand(N)
y = np.random.rand(N)

df = pd.DataFrame({'x': x, 'y': y})

ax_scatter.scatter(df.x,df.y)
ax_hist_x.hist(df.x, orientation = 'horizontal')
ax_hist_y.hist(df.y)

plt.show()
```
