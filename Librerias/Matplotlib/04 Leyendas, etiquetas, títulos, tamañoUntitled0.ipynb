```python
import matplotlib.pyplot as plt
import numpy as np
```

Creamos los datos que vamos a utilizar


```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

Graficos sin leyendas y nombres en los ejes

```python
fig, axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

Se agrega titulo

```python
fig, axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
#funcion de los titulos de grafico
axes[1].set_title('Relación X - Y')
axes[0].set_title('Relación X - Y')
plt.tight_layout()
```

Se agrega titulos a los ejes

```python
fig, axes=plt.subplots(1,2,figsize=(10,5))#figsize=(10,5) entrega el tamaño 
axes[0].plot(x,y, label="$sin(x)$")#simbolo $ se puede dejar una anotacion matematica en formati lateX
axes[1].plot(y,x)
axes[1].set_title('Relación X - Y')
axes[0].set_title('Relación X - Y')
#agrego leyendas 
axes[0].legend()
#funcion de los titulos del eje
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
plt.tight_layout()
```

```python
plt.plot(x,y, label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(0.1,0.1))#loc te indica la ubicacion de la leyenda en el grafico 
#ubicacion de la leyenda
plt.show()
```
