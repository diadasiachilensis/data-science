```python
import matplotlib.pyplot as plt
import numpy as np
```

Creamos los datos que vamos a utilizar



```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

Crear una grafica que sea de tipo fig


```python
#crea un lienzo en blanco
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

Varias graficas de tipo fig

```python
#se crean dos graficas
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
#se crean dos graficas
fig, axes = plt.subplots(nrows=1,ncols=2)
#con indice [0] para la posicion que se ubicara el grafico en el lienzo
#objetos totlamente independientes que se pueden ingresar con ellos por medio de un indice
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

Acceder a los graficos mediante el nombre almacenado en una tupla

```python
#hay otra manera que se peude acceder por medio de una tupla
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
#se configura los objetos 
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
plt.show()
```

Generar una matriz de tipo fig, con dos filas y cuatro columnas

```python
#hay otra manera que se peude acceder por medio de una tupla
fig, axes = plt.subplots(2,4) #no es necesario colocar nrow o ncols, solo poner un numeros donde corresponda (filas, columnas)
```

```python
#hay otra manera que se peude acceder por medio de una tupla
fig, axes = plt.subplots(2,4) #no es necesario colocar nrow o ncols, solo poner un numeros donde corresponda (filas, columnas)
#se configura los objetos 
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
plt.show()
```

Alternativa para generar una matriz de tipo fig, accediendo a cada grafica mediante los nombres dados en una tupla

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8))=plt.subplots(2,4) #la tupla va asi ((filas),(columnas))
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
fig.tight_layout() #esta funcin genera un espaciado decente
```

```python
#hay otra manera que se peude acceder por medio de una tupla
fig, axes = plt.subplots(2,4) #no es necesario colocar nrow o ncols, solo poner un numeros donde corresponda (filas, columnas)
#se configura los objetos 
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2,'rD:')
axes[1,0].plot(x,np.log10(x),'c') #logaritmo en base 10
axes[1,1].plot(x,np.sqrt(x),'bx:') #raiz cuadrada
axes[1,2].plot(x,np.radians(x),'cx') #radianes
axes[1,3].plot(x,np.exp(x),'ms-') #logaritmo natural
fig.tight_layout() #esta funcin genera un espaciado decente
```
