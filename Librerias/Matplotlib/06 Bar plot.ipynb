```python
import matplotlib.pyplot as plt
import numpy as np
```

Generando datos

```python
country=['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population=[10000000,12333324,13000000,12000000,12000000]
```

Grafico de variables catagoricas

```python
plt.bar(country, population, width=0.5, color=['red','blue','brown','green','black']) #width= es el ancho de als barras, color= cambian de color intercalado
plt.xticks(np.arange(5),('India','Japón','México','Colombia','Alemania'), rotation=45) #np.arange(5) genera un bublce for del set de datos que viene despues indica entre () | la rotation es de 45° de las leyendas
#se agrega titulo, nombres de los ejes por parte del reto
plt.xlabel('Paises')
plt.ylabel('Población')
plt.title('Tamaño de la población en el año 1930')
plt.show()
```

```python
plt.barh(country,population)
plt.yticks(np.arange(5),('India','Japón','México','Colombia','Alemania'))
#RETO
plt.xlabel('Paises')
plt.ylabel('Población')
plt.title('Tamaño de la población en el año 1930')
plt.show()
```
