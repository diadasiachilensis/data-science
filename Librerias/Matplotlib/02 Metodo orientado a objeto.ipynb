```python
import matplotlib.pyplot as plt
import numpy as np
```

Creamos los datos a graficar:

```python
x=np.linspace(0,5,11)
y=x**2
```

Utilizamos el concepto de objeto para crear una grafica mas personalizable:

```python
#se crea la figura, el lienzo dodne va a quedar las graficas 
#la variable es el objeto
fig= plt.figure()
#las graficas se llaman axes
axes=fig.add_axes([0.1,0.1,0.5,0.9])
#loque se quiere hacer en esta figura
axes.plot(x,y,'b')
fig.show()
```


Podemos crear varias graficas dentro del lienzo llamando el metodo fig

```python
#el mismo codigo pero no se agregan mas figuras
fig= plt.figure()
#add_axes([ubicación de la figura sobre el lienzo (0.1,0.1)| tamaño de la figura (0.5, 0.9)])
#(largo,ancho) (0.5, 0.9)
axes=fig.add_axes([0.1,0.1,0.8,0.9])
axes2=fig.add_axes([0.2,0.55,0.4,0.3])

#relacion entre vriables y la curva
axes.plot(x,y,'b')
axes2.plot(x,y,'r')
axes2.set_facecolor('gray')
fig.show()
```
